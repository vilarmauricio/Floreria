Revisamos estructura de la base de datos mysql

```python
import mysql.connector
import pandas as pd
```

```python
# Conexion a nuestra base de datos
conn = mysql.connector.connect(
    host = 'localhost',
    user = 'root', #verificar usuario
    password = '', # revisar
    port = '3306', # Si usamos el puerto por default no hace falta
    database = 'floreria', # Nombre de la base de datos si existiera
)

print(conn)
```

```python
# Listamos tablas
cursor = conn.cursor()

# Sentencia SQL
sql = "SHOW TABLES"
cursor.execute(sql)

for tabla in cursor:
    print(tabla)
```

```python
# Revisamos campos de una tabla
cursor = conn.cursor()

# Sentencia SQL
sql = "describe caja"
cursor.execute(sql)

for campo in cursor:
    print(campo)
```

TABLA caja

```python
df = pd.read_csv('../csv/2011/flujo_caja sep-2011')
df.head()
```

```python
# Seleccionamos columnas a utilizar
# fecha, $ SALDO INICIAL, $ RETIROS, $ GASTOS, $ SALDO REAL
df = df[['fecha', '$ SALDO INICIAL', '$ RETIROS', '$ GASTOS', '$ SALDO REAL']]
df = df.rename(columns= {'fecha': 'idFecha', '$ SALDO INICIAL' : 'saldo_inicial', '$ RETIROS' : 'retiro', '$ GASTOS' : 'gasto', '$ SALDO REAL' : 'cierre_caja'})
df.head(10)
```

```python
df['idFecha'] = pd.to_datetime(df['idFecha'])
```
